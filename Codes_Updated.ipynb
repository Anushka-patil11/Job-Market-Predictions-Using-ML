```python
# Inspect the dataset
df.head()

# Select the Age column
age_data = df['Age']
```

```python
scaler = MinMaxScaler()
scaled_age = scaler.fit_transform(age_data.values.reshape(-1, 1))
```

```python
# Plot the histograms
plt.figure(figsize=(12, 6))

# Original Age Distribution
plt.subplot(1, 2, 1)
plt.hist(age_data, bins=20, edgecolor='black', color='skyblue', alpha=0.7)
plt.title('Original Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Scaled Age Distribution
plt.subplot(1, 2, 2)
plt.hist(scaled_age, bins=20, edgecolor='black', color='salmon', alpha=0.7)
plt.title('Scaled Age Distribution (Min-Max)')
plt.xlabel('Scaled Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the models and metrics
models = ['Decision Trees', 'Random Forests', 'Neural Networks', 'LSTM']
rmse = [15, 12, 10, 8]  # Example RMSE values
precision = [3, 2, 2.5, 2]  # Example Precision values
recall = [1.5, 1.8, 2.2, 2.5]  # Example Recall values
f1_score = [1, 1.5, 1.8, 2]  # Example F1-Score values

# Stacking the bars
bar_width = 0.6
indices = np.arange(len(models))

# Plot the bars
plt.barh(indices, rmse, bar_width, label='RMSE', color='skyblue')
plt.barh(indices, precision, bar_width, left=rmse, label='Precision', color='lightgreen')
plt.barh(indices, recall, bar_width, left=np.array(rmse) + np.array(precision), label='Recall', color='salmon')
plt.barh(indices, f1_score, bar_width, left=np.array(rmse) + np.array(precision) + np.array(recall), label='F1-Score', color='orange')

# Add labels, title, and legend
plt.xlabel('Scores', fontsize=12)
plt.ylabel('Models', fontsize=12)
plt.title('Comparative Performance of Machine Learning Models (Horizontal Bar Graph)', fontsize=14)
plt.yticks(indices, models)
plt.legend()

# Add grid for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the models and metrics
models = ['ARIMA', 'Linear Regression', 'Random Forests', 'LSTM']
rmse = [18, 17, 10, 8]  # Example RMSE values
precision = [2, 2.5, 3, 3.5]  # Example Precision values
recall = [1, 1.2, 2, 2.2]  # Example Recall values
f1_score = [0.8, 1, 1.5, 2]  # Example F1-Score values

# Stacking the bars
bar_width = 0.6
indices = np.arange(len(models))

# Plot the bars
plt.bar(indices, rmse, bar_width, label='RMSE', color='skyblue')
plt.bar(indices, precision, bar_width, bottom=rmse, label='Precision', color='lightgreen')
plt.bar(indices, recall, bar_width, bottom=np.array(rmse) + np.array(precision), label='Recall', color='salmon')
plt.bar(indices, f1_score, bar_width, bottom=np.array(rmse) + np.array(precision) + np.array(recall), label='F1-Score', color='orange')

# Add labels, title, and legend
plt.xlabel('Models', fontsize=12)
plt.ylabel('Scores', fontsize=12)
plt.title('Comparative Performance: Traditional vs. Machine Learning Models', fontsize=14)
plt.xticks(indices, models)
plt.legend()

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
